```python
import pandas as pd
import os

data = pd.read_csv('weather_data/1. Weather Data.csv')
```

```python
# 1. Find all records where the weather was exactly clear
clear_weather = data[data['Weather_Condition'] == 'Clear']

# 2. Find the number of times the wind speed was exactly 4 km/hr
wind_speed_4 = data[data['Wind Speed_km/h'] == 4].shape[0]

# 3. Check if there are any NULL values present in the dataset
null_values = data.isnull().sum().sum()

# 4. Rename the column "Weather" to "Weather_Condition"
data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)

# 5. What is the mean visibility of the dataset?
mean_visibility = data['Visibility_km'].mean()

# 6. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km
wind_speed_visibility = data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)].shape[0]

# 7. What is the mean value of each column for each weather condition?
mean_values_per_weather = data.groupby('Weather_Condition').mean(numeric_only=True)

# 8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40
complex_condition = data[(data['Weather_Condition'] == 'Clear') & (data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40)]

# 9. Find the number of weather conditions that include snow
snow_conditions = data['Weather_Condition'].str.contains('Snow').sum()
```

```python
# Display results
print("All records where the weather was exactly clear:\n", clear_weather)
print("\nNumber of times the wind speed was exactly 4 km/hr:", wind_speed_4)
print("\nAre there any NULL values present in the dataset?", null_values > 0)
print("\nMean visibility of the dataset:", mean_visibility)
print("\nNumber of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km:", wind_speed_visibility)
print("\nMean value of each column for each weather condition:\n", mean_values_per_weather)
print("\nAll instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40:\n", complex_condition)
print("\nNumber of weather conditions that include snow:", snow_conditions)
```
